```python
max_power=[]
for i in range(len(new_df)):
    #print(i)
    speed=new_df.loc[i][11].split(" ")
    if '.' in speed[1]:
        power=re.findall("\d+\.\d+", str(speed[1]))
        max_power.append(float(power[0]))
        #print(float(power[0]))
    else:
        power = float(re.findall(r'\d+',str(speed[1]) )[0])
        max_power.append(power)
        #print(p)
        
#print(max_power)        
```

```python
new_df['max_power'] = max_power
```

```python
#New Car (On-Road Price) : Rs.10.14-13.79 Lakh*
new_price_arr=[]
for i in range(len(new_df)):
    #print(i)
    if str(new_df.loc[i][2]) =="0":
        #print(float(new_df.loc[i][2]))
        new_price_arr.append(float(new_df.loc[i][2]))
    else:
        price_string=new_df.loc[i][2].split(":")
        new_price=re.findall("\d+\.\d+",price_string[1])
        #print(float(new_price[0]))
        new_price_arr.append(float(new_price[0]))       
```

```python
new_df['new-price'] = new_price_arr
```
